```python
# dataset from https://oracleselixir.com/
# https://drive.google.com/drive/u/1/folders/1gLSw0RLjBbtaNy0dgnGQDAZOHIgCe-HH

import pandas as pd

file_name = '2023_LoL_esports_match_data_from_OraclesElixir.csv'

df = pd.read_csv(file_name)

df
```

```python
print(list(df.keys()))
```

```python
# removing incomplete data
df = df.loc[df['datacompleteness'] == 'complete']
df
```

```python
# remove information about team
df = df.loc[df['position'] != 'team']
df
```

```python
df = df[['patch', 'side', 'position', 'playername', 'champion', 'gamelength', 'result', 'kills', 'deaths', 'assists', 'firstbloodkill', 'firstbloodassist', 'firstbloodvictim', 'ckpm', 'barons', 'opp_barons', 'inhibitors', 'opp_inhibitors', 'dpm', 'damagetakenperminute', 'damagemitigatedperminute', 'wpm', 'wcpm', 'vspm', 'earned gpm', 'cspm']]
df
```

```python
df = df.dropna(axis = 0)
df
```

```python
factorize_list = ['side', 'position', 'playername', 'champion']
values_list = []
for feature in factorize_list:
    df_feature, value_list = pd.factorize(df[feature])
    df.loc[:, feature] = df_feature
    values_list.append(value_list)

for feature_no in range(len(factorize_list)):
    print(f"{factorize_list[feature_no]}: {values_list[feature_no]}")

df
```

```python
output_file_name = 'preprocessed_' + file_name
df.to_csv(output_file_name, index=False)
```
